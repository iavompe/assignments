# CTR calc
# get only bidded auctions

```python
df_ctr = df.loc[df['bidded'] == 1.0]
df_ctr = df_ctr.drop('bidded', axis=1)
df_ctr
```

```python
numeric_features_ctr = df_ctr.select_dtypes(include=np.number).columns.tolist()
categorial_features_ctr = df_ctr.select_dtypes(include=['object']).columns.tolist()

df_numeric_ctr = df[numeric_features_ctr]
df_categorial_ctr = df[categorial_features_ctr]
```

```python
sns.heatmap(df_numeric_ctr.corr())
```

```python
cnt = df_ctr['conversion'].count()
sum = df_ctr['conversion'].sum()

ctr_avg = sum / cnt * 100.00
ctr_avg
```

```python
df1 = df_ctr.groupby(['state_code'], as_index=False)['conversion'].sum()
df2 = df_ctr.groupby(['state_code'], as_index=False)['state_code'].count()

df_tmp = df1.join(df2, lsuffix="", rsuffix="_count")
df_tmp["conv_perc"] = df_tmp["conversion"] / df_tmp["state_code_count"] * 100.00
df_tmp["ctr_avg"] = ctr_avg
df_tmp.sort_values(by=['state_code_count'], ascending=False)
```

```python
df1 = df_ctr.groupby(['device_type'], as_index=False)['conversion'].sum()
df2 = df_ctr.groupby(['device_type'], as_index=False)['device_type'].count()

df_tmp = df1.join(df2, lsuffix="", rsuffix="_count")
df_tmp["conv_perc"] = df_tmp["conversion"] / df_tmp["device_type_count"] * 100.00
df_tmp["ctr_avg"] = ctr_avg
df_tmp.sort_values(by=['device_type_count'], ascending=False)

# phone works better than pc
```

With enough historical data we can use a lot of classic ML tools to create some models.

Basic steps:
1. Data Collection:
Gather historical data on bids, wins, and relevant features for each URL and ad slot combination.
Explore and clean the data to handle missing values, outliers, and ensure consistency.

2. Feature Engineering:
Identify relevant features that might influence the bid's success rate. This could include information about the user, time of day, device type, historical performance, etc.
Encode categorical variables and normalize numerical variables.

3. Split Data:
Divide the dataset into training and testing sets. This is crucial to evaluate the model's performance on unseen data.

4. Model Selection:
Choose a suitable machine learning algorithm like linear regression, decision trees, random forests, gradient boosting, etc.

5. Model Training:
Train the model on the training dataset. Use cross-validation to optimize hyperparameters and avoid overfitting.

6. Evaluation:
Evaluate the model's performance on the testing dataset using appropriate metrics (e.g., Mean Absolute Error, Mean Squared Error, etc.).
Adjust the model or try different algorithms if needed.

7. Deployment:
Once satisfied with the model performance, deploy it to a production environment where it can be used to predict bids.

8. Monitoring and Updating:
Periodically retrain the model with new data to ensure its accuracy over time.
Monitor the model's performance in the production environment and update it if necessary.
